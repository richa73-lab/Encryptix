```python
df['Age'].fillna(df['Age'].median(), inplace=True)  
```

```python
df.isnull().sum()
```

```python
# here target variabe is= survived and feaature variable are the column all excluding survive column
x = df.drop('Survived', axis=1)
y = df['Survived']

#we will split the data into (x,y)training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)#testing 80% and training 20%

# Train a Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# # Make predictions
y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')




```
